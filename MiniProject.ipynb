```python
# pip install pyttsx3

import pyttsx3

def tts(text, v):
    
    t1 = pyttsx3.init()                                         
    #Rate
    rate = t1.getProperty('rate')
    t1.setProperty('rate',115)                 #115 words per minute
    #Volume
    volume = t1.getProperty('volume')
    t1.setProperty('volume',1.0)               #highest volume
    #Voice
    voices = t1.getProperty('voices')
    t1.setProperty('voice',voices[v].id)       #id=1 is for female voice
    
    t1.say(text)                               #play the text
    t1.runAndWait()                            #returns only after all commands queued before this call are emptied
        
        
    
    
msg = """
    
                 **************** CHOOSE FROM MENU ****************
                
                 a) Press 1 to convert text to voice
                 b) Press any other number to exit
                 c) To quit the program after running, enter "exit"
                 
                 **************************************************
    """
print(msg)

n= int(input ("Enter your choice:"))

while n == 1:
    text = input ( "Enter text to be converted to voice:" )
    voice = input ("Do you want the voice in male(M) or female(F)? \n By default it is female.  ")
    if voice == "M":
        v = 0
    else:
        v = 1
    
    if text=="exit":
        print("exiting...")
        tts("exiting the program")
        break
    else:
        tts(text,v)
```
